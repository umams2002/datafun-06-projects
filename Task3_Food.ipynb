Section 9-Map

```python
def getsugarcontent(sugar_content):
    if sugar_content < 5 :
        return 'Low'
    elif 5 <= sugar_content < 8:
        return 'Medium'
    else:
        return 'High'
    
df['sugar_category'] = df['SugarTotal'].map(getsugarcontent)
df
```
